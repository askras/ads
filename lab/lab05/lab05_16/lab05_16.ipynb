## –ó–∞–¥–∞–Ω–∏–µ 2.

–î–∞–Ω –ª–∏–±–∏—Ä–∏–Ω—Ç. –í—Ö–æ–¥ –≤ –ª–∞–±–∏—Ä–∏–Ω—Ç –Ω–∞—Ö–æ–¥–∏—Ç—Å—è —Å–≤–µ—Ä—Ö—É —Å–ª–µ–≤–∞, –≤—ã—Ö–æ–¥ - —Å–Ω–∏–∑—É —Å–ø—Ä–∞–≤–∞. 
–†–µ–∞–ª–∏–∑–æ–≤–∞–≤ –∞–ª–≥–æ—Ä–∏—Ç–º –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –∑–∞–¥–∞–Ω–∏–µ–º –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã–π –ø—É—Ç—å —á–µ—Ä–µ–∑ –ª–∞–±–∏—Ä–∏–Ω—Ç.

| –í–∞—Ä–∏–∞–Ω—Ç | –õ–∞–±–∏—Ä–∏–Ω—Ç | –ê–ª–≥–æ—Ä–∏—Ç–º |
|:---|:---|:---|
| 16  | ![](../img/maze10/maze10_16.png) | –î–≤—É—Ö–ª—É—á–µ–≤–æ–π –∞–ª–≥–æ—Ä–∏—Ç–º |

```python
import copy

from PIL import Image

def print_way(maze, A_beam):
    print("__________________________________________")
    for i in range(len(maze)):
        for j in range(len(maze[i])):
            if tuple([i, j]) in A_beam:
                print(u"\u2B1C", end="")
            elif maze[i][j]:
                print("üî≥", end="")
            else:
                print(u"\u2B1B", end="")
        print()


# –û—Ç–∫—Ä—ã–≤–∞–µ–º –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ
img = Image.open("../img/maze10/maze10_16.png")
image = img.resize((21, 21))
# image.show()
# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤ —á–µ—Ä–Ω–æ-–±–µ–ª–æ–µ
bw_image = image.convert("L")

# –ü–æ–ª—É—á–∞–µ–º —Ä–∞–∑–º–µ—Ä –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
width, height = bw_image.size

# –°–æ–∑–¥–∞–µ–º –º–∞—Ç—Ä–∏—Ü—É —Ä–∞–∑–º–µ—Ä–æ–º —Å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ, –∑–∞–ø–æ–ª–Ω–µ–Ω–Ω—É—é False
matrix = [[False for _ in range(width)] for _ in range(height)]

# –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ –∫–∞–∂–¥–æ–º—É –ø–∏–∫—Å–µ–ª—é –≤ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–∏
for y in range(height):
    for x in range(width):
        # –ü–æ–ª—É—á–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ —è—Ä–∫–æ—Å—Ç–∏ –ø–∏–∫—Å–µ–ª—è (–æ—Ç 0 –¥–æ 255)
        pixel = bw_image.getpixel((x, y))

        # –ï—Å–ª–∏ —è—Ä–∫–æ—Å—Ç—å –±–æ–ª—å—à–µ –ø–æ–ª–æ–≤–∏–Ω—ã –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π, —Ç–æ –ø–∏–∫—Å–µ–ª—å —Å—á–∏—Ç–∞–µ–º –±–µ–ª—ã–º (True)
        if pixel > 127:
            matrix[y][x] = True

print_way(matrix, [])

# –¥–≤—É—Ö–ª—É—á–µ–≤–æ–π –ø–æ–∏—Å–∫

maze = copy.deepcopy(matrix)
A = tuple([1, 0])
B = tuple([19, 20])
# 4 threads mb

def go_beam(maze, point, point_2, x, y, beams_1 = None):
    A_beam = []
    A_beam.append(point)
    result_beams = []
    for _ in range(21 * 21):
        x_coord, y_coord = A_beam[-1]
        if tuple([x_coord, y_coord]) == point_2:
            print_way(maze, A_beam)
            return True
        elif beams_1:
            for beam in beams_1:
                if tuple([x_coord, y_coord]) in beam:
                    res = A_beam + beam
                    print_way(maze, res)
                    return True
        if maze[x_coord][y_coord + y]:
            A_beam.append(tuple([x_coord, y_coord + y]))
        elif maze[x_coord + x][y_coord]:
            A_beam.append(tuple([x_coord + x, y_coord]))
        else:
            print_way(maze, A_beam)
            result_beams.append(A_beam)
            flag = False
            for _ in range(len(A_beam)):
                bad_x, bad_y = A_beam.pop()
                maze[bad_x][bad_y] = False
                for _ in range(21 * 21):
                    x_coord, y_coord = A_beam[-1]
                    if maze[x_coord + x][y_coord]:
                        A_beam.append(tuple([x_coord + x, y_coord]))
                        flag = True
                    elif maze[x_coord][y_coord + y]:
                        A_beam.append(tuple([x_coord, y_coord + y]))
                        flag = True
                    else:
                        if flag:
                            print_way(maze, A_beam)
                            result_beams.append(A_beam)
                            flag = False
                        break
            break
    return result_beams


res = go_beam(maze, A, B, 1, 1)
if res is list:
    go_beam(maze, B, A, -1, -1, beams_1=res)
else:
    go_beam(maze, B, A, -1, -1)
```
